```python
import pandas as pd
```

```python
data1 = pd.read_csv('yellow_2019.csv')
data2 = pd.read_csv('green_2019.csv')
data3 = pd.read_csv('fhv_2019.csv')
complete_data_2019 = data1.merge(data2,on='Date').merge(data3,on='Date')
complete_data_2019
```

```python
complete_data_2019.to_csv("complete_data_2019.csv")
```
