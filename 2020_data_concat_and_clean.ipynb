```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
file = "combined_csv.csv"
```

```python
df_2020_less = df_2020.drop([ "tripduration","start station id", "start station latitude", "start station longitude", "end station id", "end station latitude", "end station longitude", "bikeid", "usertype", "birth year", "gender" ], axis=1)
df_2020_less.describe()
df_2020_less.dtypes
```

```python
df_2020_clean = df_2020_less[(df_2020_less['starttime'] < '2020-07-01') & (df_2020_less['starttime'] >= '2020-03-01')]
df_2020_clean
```

```python
df_2020_clean.sort_values(by=['starttime'])
```

```python
df_2020_clean['trips'] = np.arange(len(df_2020_clean))
df_2020_clean
```

```python
#df = df[["C", "A", "B"]]

citiclean = df_2020_clean[["trips","starttime","stoptime","start station name", "end station name"]]
citiclean
```

```python
citiclean.sort_values(by=['starttime'])
```

```python
#df.rename(columns={"A": "a", "B": "c"})
newcitidf=citiclean.rename(columns={"starttime":"Date", "stoptime": "end date"})
```

```python
newcitidf['Date'] = pd.to_datetime(newcitidf['Date'])
newcitidf['end date'] = pd.to_datetime(newcitidf['end date'])
```
