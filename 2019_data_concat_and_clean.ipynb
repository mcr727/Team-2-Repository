```python
import os
import glob
import pandas as pd
os.chdir("Data/2019")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
file = "combined_csv.csv"
```

```python
df_2019 = pd.read_csv(file)
df_2019
```

```python
df_2019_less = df_2019.drop(["C/A", "UNIT", "SCP", "STATION", "LINENAME", "DIVISION", "TIME", "DESC", "EXITS                                                               "], axis=1)
df_2019_less.describe()
df_2019_less.dtypes
```

```python
df_2019_less['DATE']= pd.to_datetime(df_2019_less['DATE']) 
df_2019_less.dtypes
```

```python
df_2019_clean = df_2019_less[(df_2019_less['DATE'] < '2019-07-01') & (df_2019_less['DATE'] >= '2019-03-01')]
df_2019_clean
```

```python
entries_by_date = df_2019_clean.groupby(df_2019_clean['DATE'])['ENTRIES'].sum().reset_index()
entries_by_date.head(20)
entries_by_date.to_csv('../Cleaned/entries_by_date_2019.csv')
```
